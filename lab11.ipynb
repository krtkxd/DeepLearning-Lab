```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import kagglehub

import tensorflow as tf
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, UpSampling2D,
                                     Concatenate, Dropout)
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam


# ============================================================
# 0. DOWNLOAD DATASET FROM KAGGLEHUB
# ============================================================

path = kagglehub.dataset_download("tuanledinh/processedoriginalmonuseg")
print("Path to dataset files:", path)

# Inspect the contents of the downloaded directory to understand its structure
print("Contents of downloaded dataset root:", os.listdir(path))

# The dataset usually contains:
# path/images/
# path/masks/

# Adjust IMAGE_DIR and MASK_DIR based on the actual dataset structure if necessary
# Correcting the path to include the 'processed-original-monuseg' subdirectory
DATA_SUBDIR = os.path.join(path, "processed-original-monuseg")

# Add this line to inspect the contents of DATA_SUBDIR
print("Contents of DATA_SUBDIR:", os.listdir(DATA_SUBDIR))

# Assuming images and masks are within 'train_folder' inside DATA_SUBDIR
TRAIN_FOLDER_PATH = os.path.join(DATA_SUBDIR, "train_folder")

# Inspect the contents of TRAIN_FOLDER_PATH
print("Contents of TRAIN_FOLDER_PATH:", os.listdir(TRAIN_FOLDER_PATH))

# Correcting IMAGE_DIR and MASK_DIR based on the contents of TRAIN_FOLDER_PATH
IMAGE_DIR = os.path.join(TRAIN_FOLDER_PATH, "img")
MASK_DIR  = os.path.join(TRAIN_FOLDER_PATH, "labelcol")

print("Images Folder:", IMAGE_DIR)
print("Masks Folder :", MASK_DIR)


# ============================================================
# GLOBAL SETTINGS
# ============================================================

IMG_HEIGHT = 256
IMG_WIDTH  = 256
IMG_CHANNELS = 3


# ============================================================
# HELPER FUNCTIONS
# ============================================================

def list_image_files(folder):
    exts = ("png", "jpg", "jpeg", "bmp", "tif")
    files = []
    if os.path.exists(folder):
        for ext in exts:
            files.extend([os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(ext)])
    else:
        print(f"[WARNING] Directory not found: {folder}")
    return sorted(files)

def normalize_image(img):
    return img.astype(np.float32) / 255.0

def ensure_single_channel(msk):
    if len(msk.shape) == 3:
        msk = cv2.cvtColor(msk, cv2.COLOR_BGR2GRAY)
    msk = np.expand_dims(msk, axis=-1)
    return msk


# ============================================================
# LOAD IMAGES + MASKS
# ============================================================

def load_data(image_dir, mask_dir):
    image_files = list_image_files(image_dir)
    mask_files  = list_image_files(mask_dir)

    # Map mask names for matching
    mask_map = {os.path.splitext(os.path.basename(m))[0]: m for m in mask_files}

    images, masks = [], []

    for img_path in image_files:
        base = os.path.splitext(os.path.basename(img_path))[0]
        msk_path = mask_map.get(base)

        if msk_path is None:
            print("[WARNING] No mask found for:", img_path)
            continue

        # Load and preprocess image
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
        img = normalize_image(img)

        # Load and preprocess mask
        msk = cv2.imread(msk_path)
        msk = ensure_single_channel(msk)
        msk = cv2.resize(msk, (IMG_WIDTH, IMG_HEIGHT))
        msk = normalize_image(msk)
        msk = (msk > 0.5).astype(np.float32)

        images.append(img)
        masks.append(msk)

    return np.array(images), np.array(masks)


X, Y = load_data(IMAGE_DIR, MASK_DIR)
print("Loaded Images:", X.shape)
print("Loaded Masks:", Y.shape)


# Train / Validation Split
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)


# ============================================================
# U-NET MODEL
# ============================================================

def conv_block(x, filters):
    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    return x

def build_unet(input_shape):
    inputs = Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 64)
    p1 = MaxPooling2D((2, 2))(c1)

    c2 = conv_block(p1, 128)
    p2 = MaxPooling2D((2, 2))(c2)

    c3 = conv_block(p2, 256)
    p3 = MaxPooling2D((2, 2))(c3)

    c4 = conv_block(p3, 512)
    p4 = MaxPooling2D((2, 2))(c4)

    # Bottleneck
    c5 = conv_block(p4, 1024)

    # Decoder
    u6 = UpSampling2D((2, 2))(c5)
    u6 = Concatenate()([u6, c4])
    c6 = conv_block(u6, 512)

    u7 = UpSampling2D((2, 2))(c6)
    u7 = Concatenate()([u7, c3])
    c7 = conv_block(u7, 256)

    u8 = UpSampling2D((2, 2))(c7)
    u8 = Concatenate()([u8, c2])
    c8 = conv_block(u8, 128)

    u9 = UpSampling2D((2, 2))(c8)
    u9 = Concatenate()([u9, c1])
    c9 = conv_block(u9, 64)

    outputs = Conv2D(1, 1, activation='sigmoid')(c9)

    return Model(inputs, outputs)


model = build_unet((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))
model.summary()


# ============================================================
# COMPILE MODEL
# ============================================================

model.compile(optimizer=Adam(1e-4), loss="binary_crossentropy", metrics=["accuracy"])


# ============================================================
# TRAIN MODEL
# ============================================================

history = model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=20,
    batch_size=8
)


# ============================================================
# SHOW PREDICTIONS
# ============================================================

def show_pred(idx):
    img = X_val[idx]
    true_mask = Y_val[idx]
    pred_mask = model.predict(np.expand_dims(img, 0))[0]
    pred_mask = (pred_mask > 0.5).astype(np.uint8)

    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.title("Input Image")
    plt.imshow(img)
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Ground Truth Mask")
    plt.imshow(true_mask.squeeze(), cmap="gray")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Predicted Mask")
    plt.imshow(pred_mask.squeeze(), cmap="gray")
    plt.axis("off")

    plt.show()


show_pred(0)
show_pred(1)
show_pred(2)
```

```python
import tensorflow.keras.backend as K

def dice_coefficient(y_true, y_pred, smooth=1e-6):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coefficient(y_true, y_pred)

def bce_dice_loss(y_true, y_pred):
    return tf.keras.losses.binary_crossentropy(y_true, y_pred) + dice_loss(y_true, y_pred)
```

```python
def conv_block(x, filters):
    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    return x

def build_unet(input_shape):
    inputs = Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 64);  p1 = MaxPooling2D((2, 2))(c1)
    c2 = conv_block(p1, 128);    p2 = MaxPooling2D((2, 2))(c2)
    c3 = conv_block(p2, 256);    p3 = MaxPooling2D((2, 2))(c3)
    c4 = conv_block(p3, 512);    p4 = MaxPooling2D((2, 2))(c4)

    # Bottleneck
    c5 = conv_block(p4, 1024)

    # Decoder
    u6 = UpSampling2D((2,2))(c5);  u6 = Concatenate()([u6, c4]);  c6 = conv_block(u6, 512)
    u7 = UpSampling2D((2,2))(c6);  u7 = Concatenate()([u7, c3]);  c7 = conv_block(u7, 256)
    u8 = UpSampling2D((2,2))(c7);  u8 = Concatenate()([u8, c2]);  c8 = conv_block(u8, 128)
    u9 = UpSampling2D((2,2))(c8);  u9 = Concatenate()([u9, c1]);  c9 = conv_block(u9, 64)

    outputs = Conv2D(1, 1, activation="sigmoid")(c9)

    return Model(inputs, outputs)

model = build_unet((IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS))
model.summary()
```
