```python
# ====================== EXP 13 : VAE on MNIST ======================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, Model

# 1) Load MNIST
(x_train, _), _ = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype("float32") / 255.0
x_train = np.expand_dims(x_train, axis=-1)   # shape: (60000, 28, 28, 1)

latent_dim = 2   # keep 2D for easy visualization

# ----------------- Build Encoder -----------------
encoder_inputs = layers.Input(shape=(28, 28, 1))
x = layers.Conv2D(32, 3, strides=2, padding="same", activation="relu")(encoder_inputs)
x = layers.Conv2D(64, 3, strides=2, padding="same", activation="relu")(x)
x = layers.Flatten()(x)
x = layers.Dense(128, activation="relu")(x)

z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)

def sampling(args):
    mean, log_var = args
    eps = tf.random.normal(shape=tf.shape(mean))
    return mean + tf.exp(0.5 * log_var) * eps

z = layers.Lambda(sampling, name="z")([z_mean, z_log_var])

encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# ----------------- Build Decoder -----------------
latent_inputs = layers.Input(shape=(latent_dim,), name="z_sampling")
x = layers.Dense(7 * 7 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((7, 7, 64))(x)
x = layers.Conv2DTranspose(64, 3, strides=2, padding="same", activation="relu")(x)
x = layers.Conv2DTranspose(32, 3, strides=2, padding="same", activation="relu")(x)
decoder_outputs = layers.Conv2D(1, 3, padding="same", activation="sigmoid")(x)

decoder = Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# ----------------- Custom VAE Model -----------------
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]   # ignore labels if present

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction = self.decoder(z, training=True)

            # Reconstruction loss (binary cross-entropy per pixel)
            recon_loss = tf.reduce_mean(
                tf.keras.losses.binary_crossentropy(data, reconstruction)
            ) * 28 * 28

            # KL divergence regularization
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
            )

            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))

        return {
            "loss": total_loss,
            "recon_loss": recon_loss,
            "kl_loss": kl_loss,
        }

# Instantiate and compile VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=tf.keras.optimizers.Adam())
vae.build((None, 28, 28, 1))
vae.summary()

# ----------------- Train VAE -----------------
vae.fit(x_train, epochs=10, batch_size=128)

# ----------------- Reconstructions -----------------
# Take first 10 images
sample_imgs = x_train[:10]
_, _, z_sample = encoder(sample_imgs, training=False)
decoded = decoder(z_sample, training=False).numpy()

plt.figure(figsize=(20, 4))
for i in range(10):
    # Original
    ax = plt.subplot(2, 10, i + 1)
    plt.imshow(sample_imgs[i].reshape(28, 28), cmap="gray")
    plt.axis("off")

    # Reconstructed
    ax = plt.subplot(2, 10, i + 11)
    plt.imshow(decoded[i].reshape(28, 28), cmap="gray")
    plt.axis("off")

plt.suptitle("VAE Reconstructions")
plt.show()

# ----------------- Random Samples from Latent Space -----------------
random_latent = tf.random.normal((10, latent_dim))
gen_imgs = decoder(random_latent, training=False).numpy()

plt.figure(figsize=(20, 2))
for i in range(10):
    ax = plt.subplot(1, 10, i + 1)
    plt.imshow(gen_imgs[i].reshape(28, 28), cmap="gray")
    plt.axis("off")
plt.suptitle("VAE Generated Digits")
plt.show()
```
